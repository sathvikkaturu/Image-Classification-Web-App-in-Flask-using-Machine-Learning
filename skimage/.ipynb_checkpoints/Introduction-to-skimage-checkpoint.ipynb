```python
import numpy as np
import skimage.io
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = skimage.io.imread('images/astronaut.jpg')
```

```python
img
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
img1 = skimage.io.imread('images/colorwheel.jpg')
plt.imshow(img1)
```

```python
img1.shape
```

```python
r = img1[:,:,0] # red matrix
g = img1[:,:,1] # green
b = img1[:,:,2] # blue
```

```python
r.shape
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,3,1)
plt.imshow(r,cmap='gray')
plt.subplot(1,3,2)
plt.imshow(g,cmap='gray')
plt.subplot(1,3,3)
plt.imshow(b,cmap='gray')
plt.show()
plt.imshow(img1)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,3,1)
plt.imshow(r,cmap='Reds')
plt.subplot(1,3,2)
plt.imshow(g,cmap='Greens')
plt.subplot(1,3,3)
plt.imshow(b,cmap='Blues')
plt.show()
plt.imshow(img1)
```

# Convert Image into Gray Scale

```python
import skimage.color
```

```python
plt.imshow(img)
```

```python
gray = skimage.color.rgb2gray(img)
```

```python
img.shape
```

```python
gray.shape
```

```python
plt.imshow(gray,cmap='gray')
```

```python
skimage.io.imsave('images/astro_gray.jpg',gray)
```
